Linear univariarte regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,5,13,25,33,46]).reshape((-1,1))
y=np.array([3,6,9,12,14,18])
```

```python
plt.scatter(x,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='red',marker='*')
print("MSE:",mse)
print("Intercept:",SLR.intercept_)
print("slope:",SLR.coef_)
```

Linear regression - using random generation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random

# Generate random datas
random.seed(42)
num_points = 50
x = np.linspace(0, 10, num_points)
y = 2 * x + 1 + np.random.normal(0, 1, num_points)

# Perform linear regression
model = LinearRegression().fit(x.reshape(-1, 1), y)
predicted_y = model.predict(x.reshape(-1, 1))

# Create a DataFrame and CSV file
data = pd.DataFrame({'x': x, 'y': y, 'predicted_y': predicted_y})
data.to_csv('linear_regression_data.csv', index=False)

# Plot the data and regression line
plt.scatter(x, y, label='Data')
plt.plot(x, predicted_y, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()

```

Linear regression - extracting data from csv file

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Read data from CSV
data = pd.read_csv('linear_regression_data.csv')

# Perform linear regression
model = LinearRegression().fit(data[['x']], data['y'])
predicted_y = model.predict(data[['x']])

# Plot the data and regression line
plt.scatter(data['x'], data['y'], label='Data')
plt.plot(data['x'], predicted_y, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

Linear regression - data from excel sheet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

data=pd.read_excel('/content/Book1_LR.xlsx')
x=data.loc[0:5,('x')]
y=data.loc[0:5,('y')]

print(x)
print(y)

a=np.array(x).reshape((-1,1))
b=np.array(y)
plt.scatter(a,b)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()

from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(a,b)
pred=SLR.predict(a)

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(b,pred)
plt.plot(a,pred,color='Red',marker='*')
print("MSE:",mse)
print("Intercept:",SLR.intercept_)
print("Slope:",SLR.coef_)
```
